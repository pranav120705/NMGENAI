```python
from transformers import MarianMTModel, MarianTokenizer

def translate_text(text, src_lang='en', dest_lang='es'):
    # Load pre-trained MarianMT model and tokenizer for the source-target language pair
    model_name = f'Helsinki-NLP/opus-mt-{src_lang}-{dest_lang}'
    model = MarianMTModel.from_pretrained(model_name)
    tokenizer = MarianTokenizer.from_pretrained(model_name)

    # Tokenize the input text
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)

    # Translate text
    translated = model.generate(**inputs)

    # Decode the translated text
    translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)

    return translated_text

# Example usage
if __name__ == "__main__":
    text_to_translate = "Hello, how are you?"
    source_language = 'en'  # English
    target_language = 'es'  # Spanish

    translated_text = translate_text(text_to_translate, source_language, target_language)
    print(f"Original text: {text_to_translate}")
    print(f"Translated text: {translated_text}")
```

```python
from transformers import MarianMTModel, MarianTokenizer

def translate_text(text, src_lang='en', intermediate_lang='hi', dest_lang='ta'):
    # Load pre-trained MarianMT model and tokenizer for the source-intermediate language pair (English to Hindi)
    model_name = f'Helsinki-NLP/opus-mt-{src_lang}-{intermediate_lang}'
    model = MarianMTModel.from_pretrained(model_name)
    tokenizer = MarianTokenizer.from_pretrained(model_name)

    # Tokenize the input text
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)

    # Translate text (English to Hindi)
    translated = model.generate(**inputs)

    # Decode the translated text (Hindi)
    translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)
    print(f"Translated to Hindi: {translated_text}")

    # Now, translate Hindi to Tamil
    model_name = f'Helsinki-NLP/opus-mt-{intermediate_lang}-{dest_lang}'
    model = MarianMTModel.from_pretrained(model_name)
    tokenizer = MarianTokenizer.from_pretrained(model_name)

    # Tokenize the Hindi text for translation to Tamil
    inputs = tokenizer(translated_text, return_tensors="pt", padding=True, truncation=True)

    # Translate text (Hindi to Tamil)
    translated = model.generate(**inputs)

    # Decode the final translated text (Tamil)
    final_translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)

    return final_translated_text

# Example usage
if __name__ == "__main__":
    text_to_translate = "Hello, how are you?"
    translated_text = translate_text(text_to_translate, src_lang='en', intermediate_lang='hi', dest_lang='ta')
    print(f"Original text: {text_to_translate}")
    print(f"Translated text: {translated_text}")
```

```python
from googletrans import Translator
from gtts import gTTS
import IPython.display as ipd

# Initialize translator
translator = Translator()

# Input text
text = "Hello, how are you?"

# Translate text to desired language (e.g., Spanish)
translated_text = translator.translate(text, src='en', dest='es').text
print(f"Translated Text: {translated_text}")

# Convert the translated text to speech
tts = gTTS(text=translated_text, lang='es')

# Save the audio file
tts.save("/content/output.mp3")

# Play the audio file
ipd.Audio("/content/output.mp3")
```
